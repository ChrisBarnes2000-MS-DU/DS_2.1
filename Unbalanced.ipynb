```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs
# we create two clusters of random points
n_samples_1 = 1000
n_samples_2 = 100
centers = [[0.0, 0.0], [2.0, 2.0]]
clusters_std = [1.5, 0.5]
X, y = make_blobs(n_samples=[n_samples_1, n_samples_2],
                  centers=centers,
                  cluster_std=clusters_std,
                  random_state=0, shuffle=False)
```

```python
clf = svm.SVC(kernel='linear', C=1.0)
clf.fit(X, y)
# fit the model and get the separating hyperplane using weighted classes
wclf = svm.SVC(kernel='linear', class_weight={1: 10})
wclf.fit(X, y)
```

```python
# Helper function to plot the best fit line,
# as well as marking the closest data points to the line,
# through dotted lines (margins)
def plot_decision_boundary(typ, X, y):
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max),np.arange(x2_min, x2_max))
    Z = typ.decision_function(np.array([xx1.ravel(), xx2.ravel()]).T).reshape(xx1.shape)
    plt.contour(xx1, xx2, Z, colors='b', levels=[-1, 0, 1], alpha=0.4, linestyles=['--', '-', '--'])
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
```

```python
# Helper function to Find the error rate
# Count_one number of 0 in y_true equal to 1 in y_pred?
# Count_two how many 1 in y is 0 in y_pred?
# Add above number divide over len(y) 
```

```python
def plot_scater(typ, X, y):
    color_ls = []
    # label = []
    # if y == 1, then the value at the same index in color_ls will be 'b'
    # otherwise, it will be 'r'
    # if y == 1, then the value at the same index in label will be 'H'
    # otherwise, it will be 'NH'

    y_num_zero = y_num_one = y_pred_num_zero = y_pred_num_one = 0
    for k in y:
        if k == 1:
            color_ls.append('b')
    #         label.append('H')
            y_num_zero += 1
        else:
            color_ls.append('r')
    #         label.append('NH')
            y_num_one += 1

    y_pred = clf.predict(X)
    for i in y_pred:
        if i == 0:
            y_pred_num_zero += 1
        else:
            y_pred_num_one += 1

    count_two = y_num_one + y_pred_num_zero            

    for k, (i,j) in enumerate(X):
        # add the datapoint to a scatter plot,
        # using the k'th color in colors_ls
        plt.scatter(i, j, c = color_ls[k])
        # 0.02 to the right and above the point,
        # add the k'th text label in label
    #     plt.text(i+0.02, j+0.02, label[k])

    # pass in the trained model and data points,
    # and plot the best fit line + margins
    plot_decision_boundary(typ, X, y)

#     plt.title("weighted")
    # plot data points + color/labels
    for k, (i,j) in enumerate(X):
        plt.scatter(i, j, c = color_ls[k])
    #     plt.text(i+0.02, j+0.02, label[k])

    return count_two / len(y)
```
